# Wine quality data processing

```python
# fetch dataset 
wine_quality = fetch_ucirepo(id=186) 
  
# data (as pandas dataframes) 
X: pd.DataFrame = wine_quality.data.features 
y: pd.DataFrame = wine_quality.data.targets 
  
# metadata 
print(wine_quality.metadata) 
  
# variable information 
print(wine_quality.variables)
```

```python
X
```

```python
y
```

```python
data = pd.concat([y, (X - X.mean()) / X.std()], axis=1)
data
```

```python
data.to_csv("/home/hqian/ResearchProjects/BaggingSAA/RealData/wine_quality.csv", index=False)
```

# Bike sharing data processing

```python
# fetch dataset 
bike_sharing = fetch_ucirepo(id=275) 
  
# data (as pandas dataframes) 
X: pd.DataFrame = bike_sharing.data.features 
y: pd.DataFrame = bike_sharing.data.targets 
  
# metadata 
print(bike_sharing.metadata) 
  
# variable information 
print(bike_sharing.variables)
```

```python
XComplete = pd.concat(
    [pd.get_dummies(X["mnth"], prefix="mnth", dtype=int), 
    pd.get_dummies(X["hr"], prefix="hr", dtype=int), 
    pd.get_dummies(X["weathersit"], prefix="weathersit", dtype=int), 
    X.loc[:, ["holiday", "workingday", "temp", "atemp", "hum", "windspeed"]]], axis=1
)
XComplete
```

```python
XComplete.iloc[:, -4:] = (XComplete.iloc[:, -4:] - XComplete.iloc[:, -4:].mean()) / XComplete.iloc[:, -4:].std()
```

```python
XComplete
```

```python
y
```

```python
yNorm = np.log(y+1)
yNorm = (yNorm - yNorm.mean()) / yNorm.std()
yNorm
```

```python
data = pd.concat([y, XComplete], axis=1)
data
```

```python
data.to_csv("/home/hqian/ResearchProjects/BaggingSAA/RealData/bike_sharing.csv", index=False)
```
